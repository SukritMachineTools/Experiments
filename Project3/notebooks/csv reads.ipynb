```python
#imports
import datetime
import pandas as pd
```

```python
#read google sheets 
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vT6HrGh7EOzzejrvzkG_TGUM_GoGVDuvlUq7UcYqHlESZX6Vv8Hvwatsp4FLdE4Nmff9z5LSG3KQFq9/pub?gid=1362009325&single=true&output=csv')
```

```python
df
```

```python
#renamed columns
df.rename(columns = {'Rejection':'M/C', 'Unnamed: 6':'CASTING',
                              'Unnamed: 7':'OTHER','Timming':'Start_Time','Unnamed: 11':'End_Time','Total Prod':'Total_Prod','Final Prod':'Final_Prod','Total Rej':'Total_Rej'}, inplace = True)
```

```python
df=df.drop(index=0)
df=df.reset_index(drop=True)
```

```python
type(df.loc[0,'Date'])
```

```python
df['Date'] = pd.to_datetime(df['Date'],format='%d-%m-%Y')  
```

```python
job='S.R. Shaft THD'
breakTime='01:30:00'
start_date= '2023-06-01'
end_date= '2023-07-03'
# mask = (df['date'] > start_date) & (df['date'] <= end_date)
mask = (df['Date'] > start_date) & (df['Date'] <= end_date)
datedf=df.loc[mask]
newdf= datedf[datedf['Job']==job]
newdf=newdf.reset_index(drop=True)
newdf = newdf.astype({'Total_Prod':'int','M/C':'int','CASTING':'int','OTHER':'int','Total_Rej':'int','Final_Prod':'int'})
newdf
```

```python
sumrow={'Total_Prod':sum(newdf['Total_Prod']),'M/C':sum(newdf['M/C']),'CASTING':sum(newdf['CASTING']),'OTHER':sum(newdf['OTHER']),'Final_Prod':sum(newdf['Final_Prod'])}
sumrow= pd.DataFrame(sumrow,index=['Total'])
newdf = pd.concat([newdf, sumrow], axis=0)
newdf
```

```python

df1 = df.copy()

# Convert the 'time_string' column to time object
df1['Start_Time'] = pd.to_datetime(df1['Start_Time'], format='%H:%M')
df1['End_Time'] = pd.to_datetime(df1['End_Time'], format='%H:%M')

# Calculate the time difference
df1['time_difference'] = df1['End_Time'] - df1['Start_Time']

df1['Start_Time'] = pd.to_datetime(df1['Start_Time'], format='%H:%M').dt.time
df1['End_Time'] = pd.to_datetime(df1['End_Time'], format='%H:%M').dt.time


# Display the DataFrame
print(df1)
```
