```python
#renamed columns
df.rename(columns = {'Rejection':'M/C', 'Unnamed: 6':'CASTING',
                              'Unnamed: 7':'OTHER','Timming':'Start Time','Unnamed: 11':'End Time'}, inplace = True)
```

```python
df=df.drop(index=0)
df=df.reset_index(drop=True)
```

```python
type(df.loc[0,'Date'])
```

```python
df['Date'] = pd.to_datetime(df['Date'],format='%d-%m-%Y')  
```

```python
job='S.R. Shaft THD'
breakTime='01:30:00'
start_date= '2023-06-01'
end_date= '2023-07-03'
# mask = (df['date'] > start_date) & (df['date'] <= end_date)
mask = (df['Date'] > start_date) & (df['Date'] <= end_date)
datedf=df.loc[mask]
newdf= datedf[datedf['Job']==job]
newdf=newdf.reset_index(drop=True)
newdf = newdf.astype({'Total Prod':'int','M/C':'int','CASTING':'int','OTHER':'int','Total Rej':'int','Final Prod':'int'})
newdf
```

```python
sumrow={'Total Prod':sum(newdf['Total Prod']),'M/C':sum(newdf['M/C']),'CASTING':sum(newdf['CASTING']),'OTHER':sum(newdf['OTHER']),'Final Prod':sum(newdf['Final Prod'])}
sumrow= pd.DataFrame(sumrow,index=[0])
newdf = pd.concat([newdf, sumrow], axis=1)
newdf
```
